```python
import polars as pl
import glob
```

```python
file = glob.glob('*.csv')[0]
```

```python
df = pl.read_csv(file, encoding = 'utf8-lossy', null_values = ['#N/A', ' $- ', '- ', ' -', '-', ' - '])
```

```python
df.sample(5)
```

Select only the columns of string type (`pl.Utf8`) and strip them

```python
df = df.with_columns([
    pl.col(pl.Utf8).str.strip()
])
```

```python
df.sample(5)
```

```python
df.columns = [' '.join(c.strip().split()) for c in df.columns]
```

```python
df.sample(5)
```

Let's remove the dollar signs:

```python
dollar_cols = df.columns[6:10]
print(dollar_cols)

df = df.with_columns([
    (pl.col(dollar_cols)
     .str.replace('\$', '')
     .str.replace_all(',', '')
     .str.replace_all('-', '')
     .str.strip())
])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.when(pl.col(pl.Utf8).str.lengths() == 0).then(None).otherwise(pl.col(pl.Utf8)).keep_name()
])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.col(dollar_cols).cast(float)
])
```

```python
df.sample(5)
```

```python
df.columns
```

```python
id_vars = df.columns[:6]
value_vars = df.columns[6:]
```

```python
df = df.melt(id_vars = id_vars, value_vars = value_vars, variable_name = 'payer_data_original', value_name = 'price')
```

```python
df.sample(5)
```

```python
df = df.filter(pl.col('price').is_not_null())
```

```python
df.sample(5)
```

```python
df = df.with_columns([
       pl.when(pl.col('payer_data_original') == 'GROSS CHARGE').then('GROSS').otherwise(
           pl.when(pl.col('payer_data_original') == 'CASH PRICE').then('CASH').otherwise(
               pl.when(pl.col('payer_data_original').str.contains('MINIMUM')).then('MIN').otherwise(
                   pl.when(pl.col('payer_data_original').str.contains('MAXIMUM')).then('MAX').otherwise(
                       pl.col('payer_data_original'))))).alias('payer')
   ])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.col('payer').str.split(' [').arr.get(0).alias('payer')
])
```

```python
df = df.with_columns([
    pl.when(pl.col('payer_data_original').str.contains('\[')).then(
        pl.col('payer_data_original').str.extract(r'\[(\d+)\]')
    ).otherwise(None).alias('plan_id')
])
```

```python
df.sample(5)
```

```python
df['payer'].unique().to_pandas()
```

```python
df =  df.with_columns([
        pl.when(pl.col('payer_data_original').str.contains('\[')).then(
            pl.when(pl.col('payer_data_original').str.contains('MCR|MMCR')).then('MEDICARE').otherwise(
                pl.when(pl.col('payer_data_original').str.contains('HMO')).then('HMO').otherwise(
                    pl.when(pl.col('payer_data_original').str.contains('PREFERRED|PPO')).then('PPO').otherwise(
                        pl.when(pl.col('payer_data_original').str.contains('MMCD|MCD')).then('MEDICAID').otherwise(
                            pl.when(pl.col('payer_data_original').str.contains('MMCD|MCD')).then('MEDICAID').otherwise(
                                pl.when(pl.col('payer_data_original').str.contains('CSHCS')).then('CSHCS').otherwise(
                                    None))))))).alias('plan_type')
    ])
```

```python
df.sample(5)
```

```python
df = df.rename({
    'Code Type': 'billing_code_type',
    'Procedure': 'procedure_code',
    'Code': 'billing_code',
    'NDC': 'ndc',
    'Rev Code': 'revenue_code',
    'Procedure Description': 'description_long',
})
```
