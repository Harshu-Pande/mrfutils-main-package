### Baylor hospital example

```python
import polars as pl
filename = '75-1837454_BAYLOR UNIVERSITY MEDICAL CENTER_standardcharges.csv'
url = 'https://www.bswhealth.com/-/media/project/bsw/sites/bswhealth/documents/patient-tools/estimate-cost-of-care/75-1837454_baylor-university-medical-center_standardcharges.csv'
df = pl.read_csv(filename, skip_rows = 4, null_values = ['-', 'N/A', ' ** ', '**'], infer_schema_length = 0)
```

```python
def view(df):
    return df.sample(3).to_pandas()
```

```python
view(df)
```

```python
df.filter(pl.col('Procedure Code') == '27801253')
```

Let's examine the columns

```python
df.columns
```

```python
# No idea what this is
df = df.drop('Gross Charge Min/Max')
```

```python
view(df)
```

```python
payer_index = df.columns.index('Gross Charge')
```

```python
payers = df.columns[payer_index:]
```

```python
df = df.with_columns([
    pl.col(payers).str.replace(',', '').cast(float)
])
```

```python
view(df)
```

```python
df = df.filter(pl.col('De-Identified Maximum Reimbursement*') > .5)
```

```python
df = df.rename({
    'Procedure Code':'procedure_code',
    'Procedure Name':'description',
    'NDC':'ndc',
    'DRG':'ms-drg',
    'APR-DRG':'apr-drg',
    'Default Rev Code':'rev_code',
    'Patient Type':'patient_class',
    'CPT / HCPCS Code':'hcpcs_cpt',
    'Service Package Type':'rate_desc',
})
```

Are there every HCPCS codes that are longer than 5 chars long?

```python
df.filter(pl.col('hcpcs_cpt').str.lengths() > 5)
```

There are!! We'll have to remember this. We need to exclude these.

```python
df = df.with_columns([
    pl.lit(filename).alias('filename'),
    pl.lit(-1).alias('billed_quantity'),
    pl.lit('na').alias('billing_class'),
    pl.lit('na').alias('modifier'),
    pl.col('rev_code').str.zfill(4).fill_null('na'),
])
```

```python
df_apr = df.filter(pl.col('apr-drg').is_not_null()).drop(['ms-drg', 'ndc'])
df_drg = df.filter(pl.col('ms-drg').is_not_null()).drop(['ndc', 'apr-drg'])
df_ndc = df.filter(pl.col('ndc').is_not_null()).drop(['ms-drg', 'apr-drg'])
```

```python
df_drg.sample(3)
```

```python
view(df_ndc)
```

```python
df = (df
      .filter(pl.col('ms-drg').is_null())
      .filter(pl.col('apr-drg').is_null())
      .filter(pl.col('ndc').is_null())
      .with_columns([
          pl.lit('').alias('code')
      ])
      .drop(['ms-drg', 'apr-drg', 'ndc'])
      .filter(pl.col('procedure_code').is_not_null())
     )
```

```python
df_apr = df_apr.with_columns([pl.lit('apr-drg').alias('code_type')])
df_drg = df_drg.with_columns([pl.lit('ms-drg').alias('code_type')])
df_ndc = df_ndc.with_columns([pl.lit('ndc').alias('code_type')])
df     = df.with_columns([pl.lit('cdm').alias('code_type')])
```

```python
df_apr = df_apr.rename({'apr-drg':'code'})
df_drg = df_drg.rename({'ms-drg':'code'})
df_ndc = df_ndc.rename({'ndc':'code'})
```

```python
df_drg = df_drg.with_columns(pl.col('code').str.zfill(3))
df_drg.sample(3)
```

```python
view(df)
```

```python
def melt_payers(df):
    id_vars = [c for c in df.columns if c not in payers]
    value_vars = payers
    df = df.melt(id_vars, value_vars, variable_name = 'payer_desc', value_name = 'rate').filter(pl.col('rate').is_not_null())
    return df
```

```python
df_apr.pipe(melt_payers)
```

That's interesting. No rates.

```python
df_drg = df_drg.pipe(melt_payers)
```

```python
df_ndc = df_ndc.pipe(melt_payers)
```

```python
df = df.pipe(melt_payers)
```

```python
view(df_drg)
```

```python
view(df_ndc)
```

```python
df.columns
```

```python
df = df.select([
    'filename',
    'rev_code',
    'description',
    'procedure_code',
    'code',
    'code_type',
    'hcpcs_cpt',
    'modifier',
    'billing_class',
    'patient_class',
    'billed_quantity',
    'payer_desc',
    'rate',
    'rate_desc',
])
```

```python
set(df.columns) == set(df_drg.columns)
```

```python
set(df.columns) == set(df_ndc.columns)
```

```python
df_drg = df_drg.select(df.columns)
df_ndc = df_ndc.select(df.columns)
```

```python
df = pl.concat([df, df_drg, df_ndc])
```

```python
view(df)
```

```python
df = df.with_columns([
    pl.when(pl.col('hcpcs_cpt').is_not_null() & (pl.col('code') == '')).then(pl.col('hcpcs_cpt')).otherwise(pl.col('code')).alias('code'),
    pl.when(pl.col('hcpcs_cpt').is_not_null()).then('hcpcs_cpt').otherwise(pl.col('code_type')).alias('code_type'),
])
```

```python
df = df.with_columns([
    pl.when((pl.col('code') == '') & (pl.col('code_type') == 'cdm') & (pl.col('procedure_code').is_not_null())).then(pl.col('procedure_code')).otherwise(pl.col('code')).alias('code')
])
```

```python
df = df.with_columns([pl.col('patient_class').str.to_lowercase()])
```

```python
df = df.with_columns(
        pl.when(pl.col('payer_desc') == 'Gross Charge').then('gross')
        .when(pl.col('payer_desc') == 'Discounted Cash Price').then('cash')
        .when(pl.col('payer_desc') == 'De-Identified Minimum Reimbursement*').then('min')
        .when(pl.col('payer_desc') == 'De-Identified Minimum Reimbursement*').then('max')
        .otherwise('payer').alias('payer_category')
)
```

```python
df = df.with_columns(
    pl.when(pl.col('payer_desc').str.contains('Aetna')).then('Aetna')
    .when(pl.col('payer_desc').str.contains('Triwest')).then('Triwest')
    .when(pl.col('payer_desc').str.contains('Superior Ambetter Exchange')).then('Superior Ambetter Exchange')
    .when(pl.col('payer_desc').str.contains('Cigna')).then('Cigna')
    .when(pl.col('payer_desc').str.contains('Humana')).then('Humana')
    .when(pl.col('payer_desc').str.contains('BCBS')).then('BCBS')
    .when(pl.col('payer_desc').str.contains('Wellmed')).then('Wellmed')
    .when(pl.col('payer_desc').str.contains('Parkland')).then('Parkland')
    .when(pl.col('payer_desc').str.contains('Molina')).then('Molina')
    .when(pl.col('payer_desc').str.contains('United')).then('United')
    .otherwise(None).alias('payer_name')
)
```

```python
df = df.with_columns(
    pl.when(pl.col('payer_desc').str.contains('HMO')).then('HMO')
    .when(pl.col('payer_desc').str.contains('PPO')).then('PPO')
    .when(pl.col('payer_desc').str.contains('Medicare')).then('medicare')
    .when(pl.col('payer_desc').str.contains('Medicaid')).then('medicaid')
    .otherwise(None).alias('plan_type')
)
```

```python
df = df.with_columns(
    pl.when(pl.col('code').is_not_null()).then(pl.col('code')).otherwise(None).alias('code_value')
)
```

```python

df = df.with_columns([
    pl.lit(url).alias('url'),
    pl.lit('75-1837454').alias('hospital_ein'),
    pl.lit('450021').alias('hospital_ccn'),
    pl.lit('2022-07-01').alias('file_last_updated'),
])
```

```python
df = df.with_columns([
    pl.col('hcpcs_cpt').fill_null('na'),
])
```

```python
df = df.filter(pl.col('hcpcs_cpt').str.lengths() <= 5)
```

```python
df = df.with_columns(
    pl.col('procedure_code').fill_null('na')
)
```

```python
df = df.unique()
```

```python
df.write_csv('baylor.csv')
```
