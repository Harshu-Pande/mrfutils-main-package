```python
import polars as pl
import glob
```

```python
file = glob.glob('*.csv')[0]
```

```python
df = pl.read_csv(file, encoding = 'utf8-lossy', null_values = ['#N/A', ' $- ', '- ', ' -', '-', ' - '])
```

```python
df.sample(5)
```

Select only the columns of string type (`pl.Utf8`) and strip them

```python
df = df.with_columns([
    pl.col(pl.Utf8).str.strip()
])
```

```python
df.sample(5)
```

```python
df.columns = [' '.join(c.strip().split()) for c in df.columns]
```

```python
df.sample(5)
```

Let's remove the dollar signs:

```python
dollar_cols = df.columns[6:10]
print(dollar_cols)

df = df.with_columns([
    (pl.col(dollar_cols)
     .str.replace('\$', '')
     .str.replace_all(',', '')
     .str.replace_all('-', '')
     .str.strip())
])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.when(pl.col(pl.Utf8).str.lengths() == 0).then(None).otherwise(pl.col(pl.Utf8)).keep_name()
])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.col(dollar_cols).cast(float)
])
```

```python
df.sample(5)
```

```python
df.columns
```

```python
id_vars = df.columns[:6]
value_vars = df.columns[6:]
```

```python
df = df.melt(id_vars = id_vars, value_vars = value_vars, variable_name = 'payer_desc', value_name = 'rate')
```

```python
df.sample(5)
```

```python
df = df.filter(pl.col('rate').is_not_null())
```

```python
df.sample(5)
```

```python
df = df.with_columns([
       pl.when(pl.col('payer_desc') == 'GROSS CHARGE').then('gross').otherwise(
           pl.when(pl.col('payer_desc') == 'CASH PRICE').then('cash').otherwise(
               pl.when(pl.col('payer_desc').str.contains('MINIMUM')).then('min').otherwise(
                   pl.when(pl.col('payer_desc').str.contains('MAXIMUM')).then('max').otherwise(
                       'payer')))).alias('payer_category')
   ])
```

```python
df.sample(5)
```

```python
df = df.with_columns([
    pl.col('payer_desc').str.split(' [').arr.get(0).alias('payer_name')
])
```

```python
df = df.with_columns([
    pl.when(pl.col('payer_desc').str.contains('\[')).then(pl.col('payer_name')).otherwise(None)
])
```

```python
df = df.with_columns([
    pl.when(pl.col('payer_desc').str.contains('\[')).then(
        pl.col('payer_desc').str.extract(r'\[(\d+)\]')
    ).otherwise(None).alias('plan_id')
])
```

```python
df.sample(5)
```

```python
df['payer_desc'].unique().to_pandas()
```

```python
df =  df.with_columns([
        pl.when(pl.col('payer_desc').str.contains('\[')).then(
            pl.when(pl.col('payer_desc').str.contains('MCR|MMCR')).then('medicare').otherwise(
                pl.when(pl.col('payer_desc').str.contains('HMO')).then('hmo').otherwise(
                    pl.when(pl.col('payer_desc').str.contains('PREFERRED|PPO')).then('ppo').otherwise(
                        pl.when(pl.col('payer_desc').str.contains('MMCD|MCD')).then('medicaid').otherwise(
                            pl.when(pl.col('payer_desc').str.contains('MMCD|MCD')).then('medicaid').otherwise(
                                pl.when(pl.col('payer_desc').str.contains('CSHCS')).then('cshcs').otherwise(
                                    None))))))).alias('plan_type')
    ])
```

```python
df.sample(5)
```
